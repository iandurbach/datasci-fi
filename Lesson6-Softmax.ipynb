## Code to compute softmax function

```r
logsumexp <- function (x) {
  y = max(x)
  y + log(sum(exp(x - y)))
}

softmax <- function (x) {
  exp(x - logsumexp(x))
}
```

## Example data

```r
example1=c(2, 1, 2, 0, 3)
```

## Compute softmax

```r
softmax(example1)
```

## Check that it sums to one

```r
sum(softmax(example1))
```
